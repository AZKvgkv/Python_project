### 1.1 散点图的绘制

```python
import matplotlib.pyplot as plt

x = [225.98, 247.07, 253.14, 457.85, 241.58, 301.01, 20.67, 288.64,
     163.56, 120.06, 207.83, 342.75, 147.9, 53.06, 224.72, 29.51,
     21.61, 483.21, 245.25, 399.25, 343.35]
y = [196.63, 203.88, 210.75, 372.74, 202.41, 247.61, 24.9, 239.34, 140.32,
     104.15, 176.84, 288.23, 128.79, 49.64, 191.74, 33.1, 30.74, 400.02,
     205.35, 330.64, 283.45]

plt.figure(figsize=(10, 4), dpi=100)
plt.scatter(x, y, s=30, c='red', marker='o', alpha=0.8)
# plt.scatter(x, y)
plt.show()
```

### 1.2 柱状图的绘制

```python
# 需求-对比每部电影的票房收入
plt.rcParams['font.sans-serif'] = ['SimHei']# 用来正常显示中文标签
# 1、准备数据
movie_names = ['速激9', '黑衣人：全球追缉', '哥斯拉2', '动物世界', '超人总动员2', '李茶的快活人生', '猛虫过江',
 '龙猫', '密室逃生', '熊出没', '阿丽塔', '大黄蜂', '流浪地球']
tickets = [46.01, 31.98, 26.94, 23.54, 22.54, 21.1, 19.98, 18.64, 18.04, 17.53, 16.3, 15.74, 14.49]
# 2、创建画布
plt.figure(figsize=(16, 4), dpi=100)
# 3、绘制柱状图
colors = ['red', 'orange', 'yellow', 'green', 'cyan', 'blue', 'purple', 'pink', 'gray', 'black', 'brown', 'olive', 'navy']
plt.bar(range(len(movie_names)), tickets, color= colors, width=0.3)
# 修改刻度
plt.xticks(range(len(movie_names)), movie_names)
plt.ylim([0, 50])
# 添加标题和标签
plt.title('2019年电影票房收入')
plt.ylabel('票房收入（亿元）')
# 添加网格
plt.grid(alpha=0.3)
# 4、显示图像
plt.show()
```

```python
# 需求-对比电影相同天数的票房收入
plt.rcParams['font.sans-serif'] = ['SimHei']# 用来正常显示中文标签
movie_name = ['雷神3：诸神黄昏', '正义联盟', '寻梦环游记']
first_day_tickets = [10587.6, 10062.5, 1275.7]
first_weekend_tickets = [36224.9, 34479.6, 11830]

plt.figure(figsize=(16, 4), dpi=100)

plt.bar(range(len(movie_name)), first_day_tickets, color='red', label='首日票房', width=0.2)
plt.bar([i + 0.2 for i in range(len(movie_name))], first_weekend_tickets, color='blue', label='首周票房', width=0.2)

plt.legend()

plt.xticks([i + 0.1 for i in range(len(movie_name))], movie_name)
plt.title('电影票房收入对比')
plt.grid(alpha=0.3)

plt.show()
```

### 1.3 直方图的绘制

```python
# 需求：电影时长分布状况
time = [131, 98, 125,135, 139, 131, 135, 139, 117, 120, 124, 123, 129, 125, 132, 120, 129, 131, 118, 121, 125, 126, 130, 129, 133, 127, 131, 115, 128, 121, 114, 115, 125, 109, 111, 116, 106, 110, 118, 112, 132, 144, 150]

plt.figure(figsize=(16, 5), dpi=100)
distance = 2
group_num = int((max(time) - min(time)) / distance)

plt.hist(time, bins=group_num, edgecolor='black', color='#FFD700', density=True)

plt.xticks(range(min(time), max(time) + 2)[::distance])

plt.title('电影时长分布状况')
plt.xlabel('时长（分钟）')
plt.ylabel('电影的数量')
plt.grid(linestyle='--', alpha=0.3)

plt.show()
```

### 1.4 饼图的绘制

```python
# 需求：不同电影的拍片占比
movie_name = ['雷神3：诸神黄昏', '正义联盟', '东方快车谋杀案', '寻梦环游记', '全球风暴', '降魔传', '追捕', '七十七天', '密战', '狂兽', '其他']
place_count = [60605, 54546, 45819, 28243, 13270, 9945, 7679, 6799, 6101, 4621, 20105]

plt.figure(figsize=(16, 5), dpi=100)

plt.pie(place_count, labels=movie_name, autopct='%1.2f%%',
        colors = ['#FFC0CB', '#BA55D3', 'r', '#9370DB', '#FFD700', '#FF69B4', '#87CEFA', '#778899', '#FF7F50', '#6495ED', '#FFE4E1'])
plt.axis('equal')

plt.title('不同电影的拍片占比')

plt.show()
```
