### 1 绘制一个简单的折线图

```python
# 展示上海一周的温度

plt.figure(figsize=(16, 6), dpi=100)

plt.plot([1, 2, 3, 4, 5, 6, 7], [10, 25, 28, 26, 22, 18, 15], marker='o', linestyle='--')

plt.title('Shanghai Week Temperature')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')

# 保存
plt.savefig('shanghai_week_temperature.png')

plt.show()
```

### 2 完善原始折线图

```python
#需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度
import random
import matplotlib.pyplot as plt
# 1、准备数据
x = range(0, 60)
y_shanghai = [random.uniform(15, 18) for i in x]
# 2、创建画布
plt.figure(figsize=(16, 6), dpi=100)
# 3、绘制折线图
plt.plot(x, y_shanghai, label='Shanghai', color='red', marker='o')

# 4、设置x、y轴刻度
x_label = ['11:{:02d}'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0, 40, 5))
# 5、添加描述信息
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')

plt.grid(True, linestyle='--', alpha=0.5)

plt.title('Shanghai 11am-12pm 1-hour Temperature Change')
plt.legend()
plt.show()
```

```python
# 需求：再添加一个城市的温度变化
# 收集到北京当天温度变化情况，温度在1度到3度
import random
import matplotlib.pyplot as plt
# 1、准备数据
x = range(0, 60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 2、创建画布
plt.figure(figsize=(16, 6), dpi=100)
# 3、绘制折线图
plt.plot(x, y_shanghai, label='Shanghai', color='red', marker='o')
plt.plot(x, y_beijing, label='Beijing', color='blue', linestyle='--')

# 4、设置x、y轴刻度
x_label = ['11:{:02d}'.format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0, 40, 5))
# 5、添加描述信息
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')

plt.grid(True, linestyle='--', alpha=0.5)

plt.title('Shanghai & Beijing 11am-12pm 1-hour Temperature Change')
# 显示图例
plt.legend()

plt.show()
```

### 3 多个坐标系显示--plt.subplots(面向对象的画图方法)

```python
# 需求：再添加一个城市的温度变化
# 收集到北京当天温度变化情况，温度在1度到3度
import random
import matplotlib.pyplot as plt
# 1、准备数据
x = range(0, 60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 2、创建画布
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6), dpi=100)
# 3、绘制折线图
axes[0].plot(x, y_shanghai, label='Shanghai', color='red', marker='o')
axes[1].plot(x, y_beijing, label='Beijing', color='blue', linestyle='--')

# 4、设置x、y轴刻度
x_label = ['11:{:02d}'.format(i) for i in x]
axes[0].set_xticks(x[::5],x_label[::5])
axes[0].set_yticks(range(0, 40, 5))
axes[1].set_xticks(x[::5],x_label[::5])
axes[1].set_yticks(range(0, 40, 5))
# 5、添加描述信息
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Temperature (°C)')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Temperature (°C)')

axes[0].grid(True, linestyle='--', alpha=0.5)
axes[1].grid(True, linestyle='--', alpha=0.5)

axes[0].set_title('Shanghai 11am-12pm 1-hour Temperature Change')
axes[1].set_title('Beijing 11am-12pm 1-hour Temperature Change')
# 显示图例
axes[0].legend()
axes[1].legend()

plt.show()
```

### 4 绘制数学函数图像

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y = 2 * x ** 2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y = 2x^2$')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
